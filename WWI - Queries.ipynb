### A. Which state province do we have the most suppliers in?
    
</a>

```python
query_1 = {
    "$lookup":{
        "from": "Suppliers_StateProvinceID_embed",
        "localField": "SupplierID",
        "foreignField": "SupplierID",
        "as": "StateProvince"
    }
}

query_2 = {"$unwind": "$StateProvince"}

query_3 = {"$group":
           {
            '_id':'$StateProvince.StateProvinceName',
            'SupplierCount':{'$sum':1}
            }
        }

sort = {'$sort':{'SupplierCount':-1}}

limit = {'$limit':1}
pipeline = [query_1, query_2, query_3,sort,limit]

r = list(db.suppliers.aggregate(pipeline))

r
```

### B. How many people have three or more OtherLanguage?
    
</a>

```python
# Set parameters to use $size function later
query_1b0 = {'$match':{"$and":[
    {'OtherLanguages':{'$exists': True}},
    {'OtherLanguages':{'$ne': '[]'}}, 
    {'OtherLanguages':{'$ne': None}}]
                    }}
                      
# split the strings, find the size
query_1b1 = {'$project':{'_id':False,'OtherLanguages':{'$size':{'$split':['$OtherLanguages',',']}}}}

# Set the criteria: greater or equal to 3
query_1b2 = {'$match': {'OtherLanguages' : {'$gte': 3}}}

# count the people
query_1b3 = {'$count' : 'Number of people who have three or more OtherLanguages'}

pipeline = [query_1b0, query_1b1, query_1b2, query_1b3]

r = db.people.aggregate(pipeline)

list(r)
```

### C. Top 10 most common OtherLanguage for people records.
    
</a>

```python
query_1 = {'$match':
           {"$and":[
               {'OtherLanguages':{'$exists': True}},
               {'OtherLanguages':{'$ne': '[]'}}, 
               {'OtherLanguages':{'$ne': None}}]
            }}

query_2 = { '$set':
           {'ext': {'$regexFindAll': { 'input': "$OtherLanguages", 'regex': "([A-Z]\w+)" } } } 
          }

query_3 = { '$group': 
           { '_id': 'null', 
            'extensions': { '$push': "$ext.match" } 
           } 
          }

query_4 = {
        "$unwind": "$extensions"
        }

query_5 = {
        "$unwind": "$extensions"
        }

query_6 = { '$group': 
           { '_id': '$extensions', 
            'count': { '$sum': 1 } 
           } 
          }

query_7 = {'$sort' : {'count':-1}}

query_8 = {'$limit' : 10}

pipeline = [query_1,query_2,query_3,query_4,query_5,query_6,query_7,query_8]


r = list(db.people.aggregate(pipeline))

r[:10]
```

### A. What is the average difference in days between OrderDate and ExpectedDeliveryDate for orders sold by (SalespersonPersonID) person with the name Jack Potter?
    
</a>

```python
query_1 = {
        "$lookup":{
           "from": "people",
           "localField": "SalespersonPersonID",
           "foreignField": "PersonID",
           "as": "SalesPersonInfo"
        }
    }


query_2 = {
    "$unwind": "$SalesPersonInfo"
}

query_3 = {'$match': 
               {'SalesPersonInfo.FullName': 'Jack Potter'}
          }


query_4 = {'$group':
                {'_id' : '$SalesPersonInfo.FullName', 
                 'averageDiff' : {'$avg' : {'$dateDiff': {'startDate': '$OrderDate',
                                            'endDate': '$ExpectedDeliveryDate',
                                            'unit': 'day',}
                                           }
                                 } 
                }
          }

limit = {'$limit':5}

pipeline = [query_1,query_2,query_3,query_4,limit]
    
result = list(db.orders.aggregate(pipeline))

result
```

### C. Which two items get ordered together the most?
    
</a>

```python
query_0 = {'$match': 
            { '$or': [ {'OrderID': 4}, {'OrderID': 5} ] }
          }

#order every order and itemID so the list of itemIDs be oredered and the combination always follow the same pattern
sort = {'$sort':
        {'OrderID' : 1,
        'StockItemID' : 1}
    } 

query_1 = {'$group': 
               {
                '_id': '$OrderID',      
                'combination': {'$push':'$StockItemID'}
               }
          }

#http://www.kamsky.org/stupid-tricks-with-mongodb/how-to-get-k-combinations-from-array-in-aggregation
query_2 = {'$project':
           {'comb':{'$reduce':{
                   'input':{'$range':[0,{'$size':"$combination"}]}, 
                   'initialValue':[], 
                   'in':{'$concatArrays':[ 
                      "$$value", 
                      {'$let':{
                         'vars':{'i':"$$this"},
                         'in':{'$map':{
                            'input':{'$range':[{'$add':[1,"$$i"]},{'$size':"$combination"}]},
                            'in':[ {'$arrayElemAt':["$combination","$$i"]}, {'$arrayElemAt':["$combination","$$this"]}] }}
                      }}
                   ]}
                }}
           }
          }

query_3 = {"$unwind": "$comb"}

query_4 = {'$group': 
               {
                '_id': '$comb',      
                'qty': {'$sum':1}
               }
          }

sort = {'$sort':
            {'qty' : -1}
            }

limit = {'$limit' : 5}

pipeline = [sort,query_1,query_2,query_3,query_4,sort,limit]

result = list(db.orderlines.aggregate(pipeline))
result
```

### D. For each customer category which 3 items have the ordered the most?
    
</a>

```python
query_0 = {"$project": 
            {'CustomerCategoryID': "$CustomerCategoryID",
             'CustomerID':'$CustomerID'
            }
        }

query_1 = {"$lookup":
           {
           "from": "orders",
           "localField": "CustomerID",
           "foreignField": "CustomerID",
           "as": "OrdersDone"
            }
        }

query_2 = {"$unwind": "$OrdersDone",
        }

query_3 = {"$unwind": "$OrdersDone.OrderItemsIDs",
        }

query_4 = {'$group':
                {
                 '_id' : {'CustomerCategoryID':'$CustomerCategoryID','itemID':'$OrdersDone.OrderItemsIDs'}, 
                 'count' : {'$sum': 1}
                } 
          }

query_5 = {'$sort':
            {'count' : -1}
          } 

query_6 = {'$group':
                {'_id' : '$_id.CustomerCategoryID', 
                 'ordersArray' : {'$push':  { 'itemID':'$_id.itemID','count':'$count' } }
                }
        } 
          

query_7 = { '$project': 
           {'CustomerCategoryID': '$_id', 
            'top3': { '$slice': [ "$ordersArray", 3 ] } } 
          }

pipeline = [query_0,query_1,query_2,query_3,query_4,query_5,query_6,query_7]

r = list(db.customers.aggregate(pipeline))

r[:3]
```

### A.  What is the name of the sales person with the largest sum of invoice values in 2013 (person whose customers paid the most money)?
    
</a>

```python
query_1 = {"$project": {
                "_id":0,
                "supplierdetails.SupplierName":1,
                "TransactionAmount":1,
                'TransactionYear': {
                    "$year":"$TransactionDate"}
                        }
           }

query_2 = {"$group": {
                "_id": {
                    "Supplier":"$supplierdetails.SupplierName",
                    "Year": '2013'                    
                    },
                "Total": {
                    "$sum":"$TransactionAmount"}
                    }
           }

sort = {"$sort": {
            "Total":-1}
       }

limit = {'$limit': 1}
pipeline = [query_1, query_2, sort, limit]

r = db.SuppliersTransactionsDetails.aggregate(pipeline)
pprint(list(r))

# Answer: Name of the Supplier: Fabrikam, Inc.
```

### B.  Who are the most common `PickedByPersonID` person names for orders done by customer `Adriana Pena`?
    
</a>

```python
query_7 = {
        "$lookup":{
           "from": "people",
           "localField":'PickedByPersonID', 
           "foreignField": "PersonID",
           "as": "People"
        }
    }

query_8 = {
    "$unwind": "$People"
}


query_9 = {
    '$project': {
        'CustomerName' : 1,
        'PickedBy_PersonName':'$People.FullName'
}
        
    }


query_10 = {
    '$group':
        {'_id': {'PickedBy_PersonName' : '$PickedBy_PersonName'},
         'Count' : {'$sum' : 1}
}
    } 

query_11 = {
    '$sort':
        {'Count' :-1}} 


pipeline = [query_7,query_8,query_9,query_10,query_11]

r = db.Customers_Orders_PickedByPersonID_embed.aggregate(pipeline)
pprint(list(r))
```

### C.  How many people have in their name the string `Sara`?
    
</a>

```python
import re

pat = re.compile(r'Sara', re.I)

query_1 = {
    '$match': {
        'FullName': {'$regex': pat}}
    }

query_2 = {
    '$group': {
        '_id': {'FullName' : '$FullName'}, 
        'Count' : {'$sum' : 1}            
    }
}


pipeline = [query_1, query_2]

r = db.people.aggregate(pipeline)

pprint(len(list(r)))
```

### D.  What are the top 10 most Common Names (Primary or Surnames) of people?
    
</a>

```python
query_1 = { '$project' : { 'name' : { '$split': ["$FullName", " "] } } }

query_2 = { '$unwind' : "$name" }

query_3 = { '$group': 
           { '_id': '$name', 
            'qty': { '$sum': 1 } 
           } 
          }

query_4 = {"$sort":  
           {"qty" : -1}
        }

query_5 = { '$match' : {'_id':{'$ne': ''}} }

limit = {'$limit' : 10}

pipeline = [query_1,query_2,query_3,query_4,query_5,limit]

r = list(db.people.aggregate(pipeline))

r
```
